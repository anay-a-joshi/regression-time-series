#**Question 3 - Question 5**

```r
# =========================================================
# Step 0: Setup and Data Import
# I will upload the dataset file in the Colab sidebar first.
# =========================================================

# Reading the dataset
hospital <- read.csv("/content/6414-hw6-s26-Hospital.csv")

# Viewing the dataset
print(hospital)
str(hospital)
summary(hospital)

# Renaming variables clearly for regression
# y  = monthly labor hours required
# x1 = monthly X-ray exposures
# x2 = monthly occupied bed days
# x3 = average length of patients' stays (days)

colnames(hospital) <- c("x1", "x2", "x3", "y")

# Confirming the dataset after renaming
print(hospital)
```

#**Question 3**

```r
# =========================================================
# Question 3
# =========================================================

# Fit the multiple linear regression model
model_q3 <- lm(y ~ x1 + x2 + x3, data = hospital)

# Display the full regression output
summary(model_q3)

# Extract and print coefficient estimates only
coef(model_q3)

# Optional Step: presenting the coefficients neatly
beta_estimates <- data.frame(
  Coefficient = names(coef(model_q3)),
  Estimate = as.numeric(coef(model_q3))
)
# print(" ")
print(beta_estimates)

# Writing the fitted regression equation clearly
cat("\nEstimated Regression Model:\n")
cat("y_hat =",
    round(coef(model_q3)[1], 4),
    ifelse(coef(model_q3)[2] >= 0, " + ", " - "), abs(round(coef(model_q3)[2], 4)), "*x1",
    ifelse(coef(model_q3)[3] >= 0, " + ", " - "), abs(round(coef(model_q3)[3], 4)), "*x2",
    ifelse(coef(model_q3)[4] >= 0, " + ", " - "), abs(round(coef(model_q3)[4], 4)), "*x3",
    "\n")
```

From my multiple linear regression analysis, I obtained the fitted model

\[
  y_hat = 1946.802 + 0.0386x_1 + 1.0394x_2 - 413.7578x_3
\]

where \(y\) represents monthly labor hours required, \(x_1\) is monthly X-ray exposures, \(x_2\) is monthly occupied bed days, and \(x_3\) is the average length of patients’ stays. Thus, the estimated coefficients are \(beta_hat_0 = 1946.802\), \(beta_hat_1 = 0.0386\), \(beta_hat_2 = 1.0394\), and \(beta_hat_3 = -413.7578\).

From my interpretation, holding the other variables constant, a 1 unit increase in monthly X-ray exposures is associated with an increase of about 0.0386 labor hours required, while a 1 unit increase in monthly occupied bed days is associated with an increase of about 1.0394 labor hours required. The coefficient of average length of stay is negative, suggesting that a 1 day increase in average patient stay is associated with a decrease of about 413.7578 labor hours, after controlling for the other predictors. I also observed that all three predictors are statistically significant at the 5% level, since their p-values are below 0.05.

Overall, I found that the regression model fits the data extremely well, with \(R^2 = 0.9961\) and \(Adjusted R^2 = 0.9952\). This means that about 99.61% of the variation in monthly labor hours required is explained by the three predictor variables in my model. In my view, this indicates a very strong linear relationship between labor hours required and the hospital activity measures included in the regression...

#**Question 4**

```r
# =========================================================
# Question 4
# =========================================================

# Standardized residuals
std_resid <- rstandard(model_q3)

# Cook's distance
cook_d <- cooks.distance(model_q3)

# Exact cutoff
n <- nrow(hospital)
cook_cutoff <- 4 / n

# Creating a table of diagnostic measures
diagnostics_q4 <- data.frame(
  Observation = 1:nrow(hospital),
  Standardized_Residual = std_resid,
  Cooks_Distance = cook_d
)

# Flagging unusual observations using exact values
diagnostics_q4$Unusual_Residual <- abs(diagnostics_q4$Standardized_Residual) > 2
diagnostics_q4$Influential_Cook <- diagnostics_q4$Cooks_Distance > cook_cutoff

# Printing exact values with higher precision
options(digits = 15)
print(diagnostics_q4)

cat("Cutoff for potentially unusual standardized residuals: |standardized residual| > 2\n")
cat("Exact cutoff for potentially influential observations (Cook's Distance): 4/n =", cook_cutoff, "\n")

# Plotting standardized residuals
plot(
  diagnostics_q4$Observation,
  diagnostics_q4$Standardized_Residual,
  type = "b",
  pch = 19,
  xlab = "Observation",
  ylab = "Standardized Residual",
  main = "Standardized Residuals by Observation"
)
abline(h = c(-2, 0, 2), lty = c(2, 1, 2), col = c("red", "black", "red"))

# Plotting Cook's distance
plot(
  diagnostics_q4$Observation,
  diagnostics_q4$Cooks_Distance,
  type = "h",
  lwd = 2,
  xlab = "Observation",
  ylab = "Cook's Distance",
  main = "Cook's Distance by Observation"
)
abline(h = cook_cutoff, lty = 2, col = "red")

# Labelling influential observations
influential_points <- which(cook_d > cook_cutoff)
if (length(influential_points) > 0) {
  text(
    influential_points,
    cook_d[influential_points],
    labels = influential_points,
    pos = 3,
    cex = 0.8
  )
}
```

To identify potentially unusual observations, I examined both the standardized residuals and Cook’s distance for all 16 observations. Using the usual cutoff of absolute standardized residual greater than 2, I found that none of the observations exceeded this threshold. The largest absolute standardized residual was for observation 12, with a value of approximately -1.9294, which is still slightly below the cutoff. Therefore, based on standardized residuals alone, I would conclude that there are no strongly unusual observations in terms of response outliers.

However, when I examined Cook’s distance, I found a different result. Using the cutoff 4/n = 4/16 = 0.25, I observed that observation 15 has Cook’s distance equal to approximately 1.3838 and observation 16 has Cook’s distance equal to approximately 1.3170. Both values are far greater than 0.25, so I would identify observations 15 and 16 as potentially influential observations. Observation 12, with Cook’s distance around 0.2015, is somewhat elevated but still below the usual cutoff, so I would not classify it as influential under this rule.

From my analysis, this means that although no hospital appears to be an extreme outlier in terms of standardized residuals, observations 15 and 16 have a disproportionately large effect on the fitted regression model. In my view, these two observations deserve closer investigation because they may strongly affect the estimated coefficients, fitted values, and overall interpretation of the model...

```r
# =========================================================
# Question 5
# =========================================================

# Installing and load car package (if needed on my Google Collab)
if (!require(car)) {
  install.packages("car")
  library(car)
}

# Displaying regression summary again
summary(model_q3)

# Extracting coefficient table
coef_table_q5 <- summary(model_q3)$coefficients
print(coef_table_q5)

# Computing VIF values
vif_q5 <- vif(model_q3)
print(vif_q5)

# Putting VIF values into a neat table
vif_table_q5 <- data.frame(
  Predictor = names(vif_q5),
  VIF = as.numeric(vif_q5)
)

print(vif_table_q5)

# Also showing correlation among predictors
predictor_corr_q5 <- cor(hospital[, c("x1", "x2", "x3")])
print(predictor_corr_q5)

# Optional step again: for displaying coefficient estimates with standard errors neatly
coef_summary_q5 <- data.frame(
  Coefficient = rownames(coef_table_q5),
  Estimate = coef_table_q5[, "Estimate"],
  Std_Error = coef_table_q5[, "Std. Error"],
  t_value = coef_table_q5[, "t value"],
  p_value = coef_table_q5[, "Pr(>|t|)"]
)

print(coef_summary_q5)
```

When I examined the estimated coefficients together with their standard errors, the main counterintuitive result I identified was the coefficient of x3, the average length of patients’ stays. Its estimated coefficient is negative, about -413.7578, which suggests that as the average length of stay increases, the monthly labor hours required decrease, holding the other variables constant. From a practical hospital operations perspective, I find this somewhat counterintuitive, because I would normally expect longer patient stays to be associated with greater labor requirements rather than lower ones. However, the standard error for x3 is about 98.5983, and the coefficient is still statistically significant, so this is not simply a case of an imprecise estimate with no signal at all.

In my view, the most likely reason for this counterintuitive sign is multicollinearity among the predictor variables. The correlation matrix shows that x1 and x2 are very highly positively correlated, with correlation about 0.9048, and x2 is also fairly strongly correlated with x3, with correlation about 0.6609. This means that the predictors share overlapping information, so once the model controls for the other variables, the individual coefficient of one variable can behave in an unexpected way, including having a sign that does not match the simple real-world intuition.

The VIF results support this conclusion. I obtained VIF(x1) = 7.8283, VIF(x2) = 11.3962, and VIF(x3) = 2.5196. From my analysis, x2 clearly shows serious multicollinearity because its VIF exceeds 10, and x1 also shows substantial multicollinearity because its VIF is well above 5. The VIF for x3 is lower, but it still indicates some correlation with the other predictors. Therefore, I would conclude that multicollinearity is present in this model, especially involving x1 and x2, and this helps explain why one of the coefficient estimates, particularly for x3, appears counterintuitive even though the overall model fit is extremely strong...

#**Question 6 - Question 9**

```r
# =============================================================
# Step 0: Setup and Data Import
# Now, I will upload the next dataset file in the Colab sidebar.
# =============================================================

# Reading the dataset
shop <- read.csv("/content/6414-hw6-s26-Shop.csv")

# Viewing the dataset
print(shop)
str(shop)
summary(shop)

# If the dataset has only one sales column, renaming it clearly
# so that I can use it consistently in the analysis
if (ncol(shop) == 1) {
  colnames(shop) <- c("Sales")
}

# Creating time index
shop$t <- 1:nrow(shop)

# Creating log-sales variable
shop$logSales <- log(shop$Sales)

# Creating month variable assuming monthly data
shop$Month <- factor(rep(1:12, length.out = nrow(shop)),
                     levels = 1:12,
                     labels = c("Jan", "Feb", "Mar", "Apr", "May", "Jun",
                                "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"))

# Confirming final dataset structure
print(shop)
str(shop)
```

#**Question 6**

```r
# =========================================================
# Question 6
# =========================================================

# Time plot of original sales
plot(
  shop$t, shop$Sales,
  type = "o",
  pch = 16,
  xlab = "Time (Month)",
  ylab = "Sales",
  main = "Time Plot of Monthly Shop Sales"
)

# Optional step again: for showing the time plot of log-sales for comparison
plot(
  shop$t, shop$logSales,
  type = "o",
  pch = 16,
  xlab = "Time (Month)",
  ylab = "log(Sales)",
  main = "Time Plot of Log Monthly Shop Sales"
)
```

From the time plot of the original sales data, I observe a clear upward trend over time together with a strong seasonal pattern. The series shows repeated peaks that occur at regular intervals, which suggests that shop sales are strongly affected by the tourist season each year. I also notice that the size of the seasonal swings becomes much larger as time increases. In the earlier months, the fluctuations are relatively small, whereas in the later months the peaks become dramatically higher. This tells me that the variation in sales is increasing as the level of sales increases.

From my analysis, this is exactly why it is necessary to take logarithms of the sales data before fitting a model. In the original scale, the seasonal variation appears to be multiplicative rather than additive, meaning that the amplitude of the fluctuations grows as the series level grows. Taking the logarithm helps stabilize the variance, reduces the effect of the extremely large peaks, and converts multiplicative seasonal behavior into a more additive pattern that is easier to model with regression. After looking at the log-sales plot, I can see that the seasonal pattern becomes more consistent over time and the variability is much more stable. In my view, the logarithmic transformation is appropriate because it makes the time series structure more regular and better suited for the regression model that will be fitted in the next question...

#**Question 7**

```r
# =========================================================
# Question 7
# =========================================================

# Creating quadratic time term
shop$t2 <- shop$t^2

# Making sure January is the base level
shop$Month <- relevel(shop$Month, ref = "Jan")

# Fitting the regression model using log(Sales)
model_q7 <- lm(logSales ~ t + t2 + Month, data = shop)

# Displaying full regression output
summary(model_q7)

# Coefficient table
coef_table_q7 <- summary(model_q7)$coefficients
print(coef_table_q7)

# Creating a neat summary table
coef_summary_q7 <- data.frame(
  Predictor = rownames(coef_table_q7),
  Estimate = coef_table_q7[, "Estimate"],
  Std_Error = coef_table_q7[, "Std. Error"],
  t_value = coef_table_q7[, "t value"],
  p_value = coef_table_q7[, "Pr(>|t|)"]
)

print(coef_summary_q7)

# Showing only statistically significant predictors at alpha = 0.10
significant_q7 <- coef_summary_q7[coef_summary_q7$p_value < 0.10, ]
cat("Predictors significant at alpha = 0.10:\n")
print(significant_q7)
```

Using the regression model for log(Sales), I found that the overall fit is very strong, with R-squared equal to 0.9548 and Adjusted R-squared equal to 0.9463. This tells me that the model explains most of the variation in the logged monthly sales data. At the significance level alpha = 0.10, the statistically significant predictors are t, t2, and all monthly dummy variables from February through December. In particular, the linear time trend t is highly significant, and the quadratic term t2 is also significant at the 10% level with p-value about 0.0783. Since January is the base month, each monthly dummy coefficient measures the seasonal effect of that month relative to January.

From my analysis, the positive coefficient of t and the positive coefficient of t2 indicate that the series has an upward trend over time, and that this growth is slightly accelerating rather than remaining purely linear. I also observe strong seasonality throughout the year. All of the monthly dummy coefficients are positive, which means that every month from February to December tends to have higher log-sales than January, after controlling for trend. The largest seasonal effects occur in November and especially December, which suggests that sales rise dramatically toward the end of the year. Months such as March, July, September, and October also show strong positive seasonal effects, while February has the smallest seasonal increase relative to January.

Overall, I would describe this time series as having both a strong upward long-term trend and a strong recurring seasonal pattern. In my view, the regression results confirm what I saw in the time plots: the shop’s sales grow over time and also rise and fall systematically according to the month of the year, with the highest sales occurring in the late-year holiday season...

No, the seasonal effect in July is not identical to the seasonal effect in January. The seasonal effect for July is measured by the coefficient of MonthJul, which is 0.610252327427819. Therefore, after controlling for the time trend and quadratic trend, log-sales in July are higher than log-sales in January by about 0.6103 on the log scale.

I also observe that the p-value for the July coefficient is 3.06576551330523e-08, which is far below alpha = 0.10. This provides very strong evidence that the seasonal effect in July is significantly different from the seasonal effect in January. From my analysis, I would conclude that July has a distinctly stronger seasonal effect than January, which is consistent with the idea that tourist activity in the resort town is much higher in July than in January...
